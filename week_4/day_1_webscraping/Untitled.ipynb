```python
page = requests.get('https://www.the-numbers.com/box-office-star-records/domestic/lifetime-specific-technical-role/director')
print(page)
page.content
web_content = BeautifulSoup(page.content, 'html.parser')
web_content
print(web_content.prettify())
```
