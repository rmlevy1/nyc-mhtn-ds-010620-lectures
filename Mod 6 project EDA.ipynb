Removing Stop Words

```python
nonwords = set((x for x in tokenized_description if len(x) <=2))
```

#### Comparing Bigrams

```python
def bigrams_not_in_common(bigram1,bigram2):
    bnic =[]
    for x in bigram1:
        if x not in bigram2:
            bnic.append(x)
    return bnic
```

```python
def bigrams_in_common(bigram1,bigram2):
    bic =[]
    for x in bigram1:
        if x in bigram2:
            bic.append(x)
    return bic
```
